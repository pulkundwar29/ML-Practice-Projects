```python
datagen = ImageDataGenerator( #rescale=1./255,
                              horizontal_flip = True,
                              validation_split=0.2,
                              rotation_range=20,
                              width_shift_range=0.1,
                              height_shift_range=0.1,
                              fill_mode='nearest')

datagen2=ImageDataGenerator( rescale=1./255, horizontal_flip = False,validation_split=0.2 )

image_train=datagen.flow_from_directory( 'Cars Dataset/train',
                                          target_size=(64,64),
                                          class_mode="categorical",
                                          shuffle = True,
                                          batch_size=16,
                                          subset='training')

image_val=datagen.flow_from_directory( 'Cars Dataset/train',
                                        target_size=(64,64),
                                        class_mode="categorical",
                                        shuffle = False,
                                        batch_size=16,
                                        subset='validation')

image_test=datagen.flow_from_directory( 'Cars Dataset/test',
                                          target_size=(64,64),
                                          class_mode="categorical",
                                          shuffle = False,
                                          batch_size=16 )
```

```python
ips = (64, 64, 3)

tm1 = Sequential()
tm1.add(Conv2D(8, (3,3), 1, activation='relu', input_shape=ips))
tm1.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
tm1.add(BatchNormalization())
tm1.add(Conv2D(16, (3, 3), 1, activation='relu'))
tm1.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
tm1.add(BatchNormalization())
tm1.add(Dropout(0.5))
tm1.add(Flatten())
tm1.add(Dense(256,activation="relu"))
tm1.add(Dense(64, activation="relu"))
tm1.add(Dense(7, activation="softmax"))

tm1.summary()
```

```python
tm1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = tm1.fit(image_train, epochs=35, validation_data=image_val)
```
